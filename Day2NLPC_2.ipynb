```python
import pandas as pd,numpy as np,matplotlib.pyplot as plt,seaborn as sns,warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r"C:\Users\Asus\Downloads\Day1_c6\ML471_S2_Datafile_Concept(in).csv")
df.head()
```

```python
# Split the data into train and test sets
train_size = int(len(df) * 0.8)
train, test = df[:train_size], df[train_size:]
```

```python
# Build AR(2) model based on PACF

from statsmodels.tsa.arima.model import ARIMA
ar_model = ARIMA(train["Power_Consumption_diff"], order=(2, 0, 0))  # (p, d, q) -> AR(2) model
ar_fitted_model = ar_model.fit()
ar_fitted_model.summary()
```

```python
# Build MA(1) model based on ACF

from statsmodels.tsa.arima.model import ARIMA
ma_model = ARIMA(train["Power_Consumption_diff"], order=(0, 0, 1))  # (p, d, q) -> MA(1) model
ma_fitted_model = ma_model.fit()
ma_fitted_model.summary()
```

Day2_C3

```python
import statsmodels.api as sm

# AIC and BIC values for AR(2) model
print("AR(2) Model:")
print(f"  AIC: {ar_fitted_model.aic:.3f}")
print(f"  BIC: {ar_fitted_model.bic:.3f}")

# Ljung-Box test for AR(2) model residuals
# Check for white noise in residuals
# Null Hypothesis (H0): The residuals are independently distributed (i.e., constitute white noise).
# Alternative Hypothesis (H1): The residuals are not independently distributed.
print("  Ljung-Box test on AR(2) residuals:")
ljung_box_ar = sm.stats.acorr_ljungbox(ar_fitted_model.resid, lags=[10], return_df=True)
print(ljung_box_ar)


print("\nMA(1) Model:")
print(f"  AIC: {ma_fitted_model.aic:.3f}")
print(f"  BIC: {ma_fitted_model.bic:.3f}")

# Ljung-Box test for MA(1) model residuals
print("  Ljung-Box test on MA(1) residuals:")
ljung_box_ma = sm.stats.acorr_ljungbox(ma_fitted_model.resid, lags=[10], return_df=True)
print(ljung_box_ma)
```
